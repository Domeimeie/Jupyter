# Neuron

## Ausgangslage

```python
import matplotlib.pyplot as plt
import random
```

```python
x_inputs = [0, 1, 2, 3, 4, 5]
y_train_inputs = [2, 2.5, 3, 3.5, 4, 4.5]
```

```python
plt.scatter(x_inputs, y_train_inputs, color='red', label='Data points', s=50)

# Plot prediction line
plt.plot(x_inputs, y_train_inputs, color='blue', linestyle='--', label='Prediction line')

# Labels, title, and legend
plt.xlabel('X inputs')
plt.ylabel('Y values')
plt.title('Data points and linear prediction')
plt.grid(True)
plt.legend()

plt.show()
```

## Bereitstellung des Modells

```python
weight = random.randint(1,10)
print(f"Random weight: {weight}")
```

```python
bias = random.randint(1,10)
print(f"Random bias: {bias}")
```

```python
epochs = 10
learningrate = 0.1
print(f"Wir werden {epochs} Mal trainieren")
```

## Start Training

```python
for epoch in range(epochs):
    print(f"Ich trainiere in Epoche {epoch+1}")
    
    package = zip(x_inputs, y_train_inputs)
    
    for x, y_true in package:
        print(f"Wir haben {x} und {y_true} als Trainings Daten")

        y_pred = (weight * x) + bias
        print(f"Ich sage y vorraus: {y_pred}")

        diff = y_pred - y_true
        print(f"Ich lag um {diff} daneben")

        #calculate error
        delta = 1.0
        absd = abs(diff)
        if absd <= delta:
            error = 0.5 * diff * diff
        else:
            error = delta * absd - 0.5 * delta * delta

        #Model learning
        adjust_weight = weight * error * x
        adjust_bias = bias * error

        weight -= learningrate * adjust_weight
        bias -= learningrate * adjust_bias

        
    print("\n")
    
```
