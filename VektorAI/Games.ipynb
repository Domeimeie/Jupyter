### Setup data

```python
import numpy as np
import pandas as pd


# 1. Liste mit 10 verschiedenen Games
games = [
"Minecraft",
"Fortnite",
"The Witcher 3",
"Among Us",
"Stardew Valley",
"Dark Souls",
"League of Legends",
"Overwatch",
"Celeste",
"Animal Crossing"
]


# 2. Bewertung der Games in drei Dimensionen (Skala 1–10)
# Accessability: Wie leicht ein Spiel zugänglich ist (Steuerung, Einstieg, Zielgruppe)
# Price: Niedrig = 1 (günstig/free), Hoch = 10 (teuer)
# Complexity: Spieltiefe und Schwierigkeitsgrad
ratings = [
[8, 5, 6], # Minecraft
[9, 2, 5], # Fortnite
[5, 8, 9], # The Witcher 3
[10, 1, 3], # Among Us
[8, 4, 6], # Stardew Valley
[3, 7, 10], # Dark Souls
[6, 0, 8], # League of Legends
[7, 6, 7], # Overwatch
[7, 2, 8], # Celeste
[9, 6, 4] # Animal Crossing
]


# 3. NumPy-Array und DataFrame erzeugen
ratings_array = np.array(ratings)
df = pd.DataFrame(ratings_array, columns=["Accessibility", "Price", "Complexity"], index=games)


# Ausgabe
print(df)


df.head()
```

### 3D Visualization

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# 3D Plot erstellen
fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection='3d')

# Punkte plotten
ax.scatter(
    df["Accessibility"], 
    df["Price"], 
    df["Complexity"], 
    c="skyblue", 
    s=80, 
    edgecolors="k"
)

# Beschriftungen der Punkte (Game-Namen)
for game, row in df.iterrows():
    ax.text(row["Accessibility"], row["Price"], row["Complexity"], game, fontsize=8)

# Achsentitel
ax.set_xlabel("Accessibility")
ax.set_ylabel("Price")
ax.set_zlabel("Complexity")
ax.set_title("Games im 3D-Vektorraum")

plt.show()
```

### Ähnlichkeiten mis COS vergleichen

```python
from numpy.linalg import norm

def cosine_similarity(vec1, vec2):
    return np.dot(vec1, vec2) / (norm(vec1) * norm(vec2))

# Beispiel 1: Minecraft vs. Stardew Valley (beide kreativ & zugänglich)
vec1 = df.loc["Minecraft"].values
vec2 = df.loc["Stardew Valley"].values
print("Minecraft vs. Stardew Valley:", cosine_similarity(vec1, vec2))

# Beispiel 2: Dark Souls vs. Celeste (beide schwer & komplex, aber anderes Preisniveau)
vec1 = df.loc["Dark Souls"].values
vec2 = df.loc["Celeste"].values
print("Dark Souls vs. Celeste:", cosine_similarity(vec1, vec2))

# Beispiel 3: Among Us vs. The Witcher 3 (sehr unterschiedliche Games)
vec1 = df.loc["Among Us"].values
vec2 = df.loc["The Witcher 3"].values
print("Among Us vs. The Witcher 3:", cosine_similarity(vec1, vec2))
```
